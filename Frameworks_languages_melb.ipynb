```r
library(ggplot2)
library(gridExtra)
```

```r
df = read.csv('melb_tech.csv')
df[is.na(df)] <- 0 
langs <- unique(df$Language)
df_lang <- df[df$Framework %in% langs,]
df_fw <- df[!df$Framework %in% langs,]
```

```r
p1 <- ggplot(df_lang, aes(x=reorder(Language,Seek), y=Seek)) +
  geom_bar(stat='identity') +
  coord_flip() + ggtitle('Adzuna') + xlab('') + ylab('')


p2 <- ggplot(df_lang, aes(x=reorder(Language,Linkedin), y=Linkedin)) +
  geom_bar(stat='identity') +
  coord_flip() + ggtitle('LinkedIn')+ xlab('')+ ylab('')

p3 <- ggplot(df_lang, aes(x=reorder(Language,Indeed), y=Indeed)) +
  geom_bar(stat='identity') +
  coord_flip()+ ggtitle('Indeed')+ xlab('')+ ylab('')


p4 <- ggplot(df_lang, aes(x=reorder(Language,Adzuna), y=Adzuna)) +
  geom_bar(stat='identity') +
  coord_flip() + ggtitle('Adzuna')+ xlab('')+ ylab('')


grid.arrange(p1, p2,p3,p4, ncol = 2, nrow=2)
```

```r
ggplot(df_fw, aes(x=reorder(Framework,Seek), y=Seek, fill=Language)) +
  geom_bar(stat='identity') +
  coord_flip()
```

```r
ggplot(df_fw, aes(area = Seek, fill = Language, label = Framework,
                subgroup = Language)) +
  geom_treemap() +
  geom_treemap_subgroup_border() +
  geom_treemap_subgroup_text(place = "centre", grow = T, alpha = 0.5, colour =
                             "black", fontface = "italic", min.size = 0) +
  geom_treemap_text(colour = "white", place = "topleft", reflow = T)
```
